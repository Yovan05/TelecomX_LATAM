```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
clientes = pd.read_json(url)
```

```python
clientes_normalizados = pd.json_normalize(clientes.to_dict(orient="records"))
clientes_normalizados
```

#ðŸ”§ TransformaciÃ³n

```python
clientes_normalizados['internet.OnlineSecurity']=clientes_normalizados['internet.OnlineSecurity'].replace('No internet service','No')
```

```python
clientes_normalizados["Churn"]=clientes_normalizados["Churn"].map({"Yes": True, "No": False})
```

```python
clientes_normalizados["Churn"]=clientes_normalizados["Churn"].astype(bool)
```

```python
clientes_normalizados.info()
```

```python
clientes_normalizados['customer.SeniorCitizen']=clientes_normalizados['customer.SeniorCitizen'].astype(bool)
```

```python
columnas=['customer.Partner','customer.Dependents','phone.PhoneService','account.PaperlessBilling']
```

```python
clientes_normalizados[columnas] = clientes_normalizados[columnas].applymap(lambda x: True if x=="Yes" else False)
```

```python
clientes_normalizados[columnas]=clientes_normalizados[columnas].astype(bool)
```

```python
clientes_normalizados.info()
```

```python
clientes_normalizados["account.Charges.Total"]=clientes_normalizados["account.Charges.Total"].str.replace(r"[^0-9.]", "", regex=True)
```

```python
num_nan = clientes_normalizados["account.Charges.Total"].isna().sum()
print("Cantidad de NaN:", num_nan)
```

```python
num_empty = (clientes_normalizados["account.Charges.Total"] == "").sum()
print("Cantidad de cadenas vacÃ­as:", num_empty)
```

```python
clientes_normalizados[clientes_normalizados['account.Charges.Total']=='']
```

```python
clientes_normalizados = clientes_normalizados[
    ~(clientes_normalizados["account.Charges.Total"].isna() |
      (clientes_normalizados["account.Charges.Total"] == ""))
]
```

```python
clientes_normalizados['account.Charges.Total']=clientes_normalizados['account.Charges.Total'].astype(np.float64)
```

```python
clientes_normalizados.info()
```

```python
clientes_normalizados['account.PaymentMethod'].value_counts()
```

```python
clientes_normalizados["customerID"].duplicated().any()
```

```python
clientes_normalizados["GastoDiario"] = clientes_normalizados["account.Charges.Monthly"] / 30
```

```python
clientes_normalizados=clientes_normalizados.rename(columns={'GastoDiario': 'gasto_diario'})
```

```python
clientes_normalizados.info()
```

```python
# Estilo de seaborn
sns.set_theme(style="whitegrid")

# Contar valores de Churn
churn_counts = clientes_normalizados['Churn'].value_counts()
churn_labels = churn_counts.index
churn_sizes = churn_counts.values

# Crear la grÃ¡fica de pastel
fig, ax = plt.subplots(figsize=(7,7))
wedges, texts, autotexts = ax.pie(
    churn_sizes,
    labels=churn_labels,
    autopct=lambda p: f"{int(p*sum(churn_sizes)/100)} ({p:.1f}%)",  # nÃºmero y porcentaje
    startangle=90,
    colors=sns.color_palette("pastel")
)

# Estilo de textos
for text in texts + autotexts:
    text.set_fontsize(12)

# TÃ­tulo
ax.set_title("DistribuciÃ³n de clientes que dejaron la empresa (Churn)", fontsize=16)

plt.show()
```

```python
sns.set_theme(style="whitegrid")

# Renombrar categorÃ­as para el grÃ¡fico
clientes_normalizados["EdadCategoria"] = clientes_normalizados["customer.SeniorCitizen"].map({True: ">=65", False: "<65"})

# Contar clientes por categorÃ­a de edad y Churn
conteo = clientes_normalizados.groupby(["EdadCategoria", "Churn"]).size().reset_index(name="Cantidad")

# Crear grÃ¡fica de barras horizontales
plt.figure(figsize=(8,6))
ax = sns.barplot(
    data=conteo,
    y="EdadCategoria",        # eje Y = categorÃ­a de edad
    x="Cantidad",             # eje X = cantidad
    hue="Churn",
    palette={True: "red", False: "green"}
)

plt.title("Clientes por categorÃ­a de edad", fontsize=16)
plt.xlabel("NÃºmero de Clientes")
plt.ylabel("CategorÃ­a de Edad")

# Personalizar leyenda
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=["No Cancelaron", "Cancelaron"])

# Agregar nÃºmeros encima de cada barra
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 0.5,          # posiciÃ³n X (al lado de la barra)
            p.get_y() + p.get_height()/2,  # posiciÃ³n Y (centro de la barra)
            int(width),          # texto = valor entero
            va='center')

plt.show()
```

Se puede observar que aunque la cantidad bruta de personas que mantiene o cancelaron la suscripcion es similar en los clientes mayores a 65 aÃ±os, el porcentaje de los clientes mayores a 65 aÃ±os que cancelaron es mayor a los que no cancelaron con respecto a los clientes menores a 65 aÃ±os, lo que indica que los clientes mayores a 65 aÃ±os tienden mas a cancelar que los menores a 65

## **Analisis por tiempo de contrato**

```python
clientes_normalizados['customer.tenure'].value_counts()
```

```python
sns.set_theme(style="whitegrid")

# Crear bins de 6 en 6 hasta 72 meses
bins = list(range(0, 73, 6))  # 0,6,12,...72
labels = [f"{b}-{b+5}" for b in bins[:-1]]  # etiquetas: "0-5", "6-11", etc.

# Crear columna de rango de tenure
clientes_normalizados["TenureGrupo"] = pd.cut(clientes_normalizados["customer.tenure"], bins=bins, labels=labels, right=True, include_lowest=True)

# Contar clientes por grupo de tenure y Churn
conteo = clientes_normalizados.groupby(["TenureGrupo", "Churn"]).size().reset_index(name="Cantidad")

# Definir colores pastel fijos
colores = {True: sns.color_palette("pastel")[1], False: sns.color_palette("pastel")[0]}

# Crear grÃ¡fica de barras verticales
plt.figure(figsize=(12,6))
ax = sns.barplot(
    data=conteo,
    x="TenureGrupo",
    y="Cantidad",
    hue="Churn",
    palette=colores
)

plt.title("Cantidad de clientes segÃºn meses de contrato y Churn", fontsize=16)
plt.xlabel("Meses de Contrato (Tenure)")
plt.ylabel("NÃºmero de Clientes")

# Personalizar leyenda
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=["No Cancelaron", "Cancelaron"], title="Churn")

# Agregar valores encima de cada barra
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, height + 0.5, int(height), ha='center')

plt.show()
```
