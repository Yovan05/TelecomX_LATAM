```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = '/content/TelecomX_Data.json'
```

```python
clientes = pd.read_json(url)
```

```python
clientes_normalizados = pd.json_normalize(clientes.to_dict(orient="records"))
clientes_normalizados
```

```python
clientes_normalizados.info()
```

#ðŸ”§ TransformaciÃ³n

```python
clientes_normalizados['internet.OnlineSecurity']=clientes_normalizados['internet.OnlineSecurity'].replace('No internet service','No')
```

```python
clientes_normalizados["Churn"]=clientes_normalizados["Churn"].map({"Yes": True, "No": False})
```

```python
clientes_normalizados["Churn"]=clientes_normalizados["Churn"].astype(bool)
```

```python
clientes_normalizados.info()
```

```python
clientes_normalizados['customer.SeniorCitizen']=clientes_normalizados['customer.SeniorCitizen'].astype(bool)
```

```python
columnas=['customer.Partner','customer.Dependents','phone.PhoneService','account.PaperlessBilling']
```

```python
clientes_normalizados[columnas] = clientes_normalizados[columnas].applymap(lambda x: True if x=="Yes" else False)
```

```python
clientes_normalizados[columnas]=clientes_normalizados[columnas].astype(bool)
```

```python
clientes_normalizados.info()
```

```python
clientes_normalizados["account.Charges.Total"]=clientes_normalizados["account.Charges.Total"].str.replace(r"[^0-9.]", "", regex=True)
```

```python
num_nan = clientes_normalizados["account.Charges.Total"].isna().sum()
print("Cantidad de NaN:", num_nan)
```

```python
num_empty = (clientes_normalizados["account.Charges.Total"] == "").sum()
print("Cantidad de cadenas vacÃ­as:", num_empty)
```

```python
clientes_normalizados[clientes_normalizados['account.Charges.Total']=='']
```

```python
clientes_normalizados = clientes_normalizados[
    ~(clientes_normalizados["account.Charges.Total"].isna() |
      (clientes_normalizados["account.Charges.Total"] == ""))
]
```

```python
clientes_normalizados['account.Charges.Total']=clientes_normalizados['account.Charges.Total'].astype(np.float64)
```

```python
clientes_normalizados.info()
```

```python
clientes_normalizados['account.PaymentMethod'].value_counts()
```

```python
clientes_normalizados["customerID"].duplicated().any()
```

```python
clientes_normalizados["GastoDiario"] = clientes_normalizados["account.Charges.Monthly"] / 30
```

```python
clientes_normalizados=clientes_normalizados.rename(columns={'GastoDiario': 'gasto_diario'})
```

```python
clientes_normalizados.info()
```

## **Analisis por tiempo de contrato**

```python
clientes_normalizados['customer.tenure'].value_counts()
```

```python
sns.set_theme(style="whitegrid")

# Crear bins de 6 en 6 hasta 72 meses
bins = list(range(0, 73, 6))  # 0,6,12,...72
labels = [f"{b}-{b+5}" for b in bins[:-1]]  # etiquetas: "0-5", "6-11", etc.

# Crear conteo
conteo = (
    clientes_normalizados
    .assign(TenureGrupo=pd.cut(clientes_normalizados["customer.tenure"], bins=bins, labels=labels, right=True, include_lowest=True))
    .groupby(["TenureGrupo", "Churn"])
    .size()
    .reset_index(name="Cantidad")
)

# Definir colores pastel fijos
colores = {True: sns.color_palette("pastel")[1], False: sns.color_palette("pastel")[0]}

# Crear grÃ¡fica de barras verticales
plt.figure(figsize=(12,6))
ax = sns.barplot(
    data=conteo,
    x="TenureGrupo",
    y="Cantidad",
    hue="Churn",
    palette=colores
)

plt.title("Cantidad de clientes segÃºn meses de contrato y Churn", fontsize=16)
plt.xlabel("Meses de Contrato")
plt.ylabel("NÃºmero de Clientes")

# Personalizar leyenda
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=["No Cancelaron", "Cancelaron"])

# Agregar valores encima de cada barra
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, height + 0.5, int(height), ha='center')

plt.show()
```

La grafica muestra claramente las 2 tendencias que existen, mientras que conforme pasas los meses, los cliententes tienden a mantener sus suscripciones, los nuevos clientes tienden a  cancelar en los primeros 5 meses. La tendencia de los clientes que se mantienen sigue la forma de una parabola invertida, mientras que la tendencia de los clientes que ya cancelaron forma una hiperbola

## **Analisis de gastos**

```python
# Definir bins para gasto_diario (puedes ajustar el tamaÃ±o del intervalo si quieres mÃ¡s detalle)
bins = pd.interval_range(start=clientes_normalizados["gasto_diario"].min(),
                         end=clientes_normalizados["gasto_diario"].max()+0.5,
                         freq=0.5, closed="left")

# Crear columna temporal con los rangos de gasto
clientes_temp = clientes_normalizados.assign(
    gasto_rango=pd.cut(clientes_normalizados["gasto_diario"], bins=bins)
)

# Contar usuarios por rango y churn
conteo = (
    clientes_temp
    .groupby(["gasto_rango", "Churn"])
    .size()
    .reset_index(name="Cantidad")
)

# Contar todos los usuarios (sin importar churn)
conteo_total = (
    clientes_temp
    .groupby(["gasto_rango"])
    .size()
    .reset_index(name="Cantidad")
)
conteo_total["Churn"] = "Todos"

# Unir los dos conteos
conteo["Churn"] = conteo["Churn"].map({True: "Cancelaron", False: "Activos"})
conteo_final = pd.concat([conteo, conteo_total])

# Redondear los rangos a 2 decimales para mostrar en el eje X
conteo_final["gasto_rango"] = conteo_final["gasto_rango"].apply(
    lambda x: f"{x.left:.2f} - {x.right:.2f}"
)

# Crear grÃ¡fica dinÃ¡mica de lÃ­neas
fig = px.line(
    conteo_final,
    x="gasto_rango",
    y="Cantidad",
    color="Churn",
    markers=True,
    title="Usuarios por gasto diario y estado de cancelaciÃ³n"
)

# Personalizar ejes
fig.update_layout(
    xaxis_title="Rango de gasto diario",
    yaxis_title="NÃºmero de usuarios",
    legend_title="Estado"
)

fig.show()
```

La unica diferencia de tendencia que se ve en la grafica es al incio, en donde se puede observar que los clientes que menos gastan, son los menos propensos a cancelar, por los demas rangos de gastos, las graficas se tornan similares
